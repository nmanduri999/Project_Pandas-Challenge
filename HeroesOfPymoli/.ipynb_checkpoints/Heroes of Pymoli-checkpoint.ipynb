```python
# Import Dependencies
import pandas as pd
```

```python
# File to Load 
file_to_load = "Resource/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load, encoding="utf-8")
purchase_data_df.head()
```

```python
purchase_data_df.columns
```

# Player Count

```python
#Total Number of Players
Total_player = len(purchase_data_df["SN"].value_counts())
player_table = pd.DataFrame({"Total player": Total_player},index =[0])
player_table
```

# Purchasing Analysis (Total)

```python
#Number of Unique Items
item_id = len(purchase_data_df["Item ID"].value_counts())
#Average Purchase Price
avg_price = purchase_data_df["Price"].mean()
#Total Number of Purchases
total_player = purchase_data_df["Purchase ID"].count()
#Total Revenue
total_revenue = purchase_data_df["Price"].sum()
#Table Format
purchase_df = pd.DataFrame({"Number of Unique Items": [item_id],
                            "Average Purchase Price": [avg_price],
                            "Number of Purchases": [total_player],
                            "Total Revenue": [total_revenue] },index =[0])

purchase_df["Average Purchase Price"] = purchase_df["Average Purchase Price"].map('${:,.2f}'.format)
purchase_df["Total Revenue"] = purchase_df["Total Revenue"].map('${:,.2f}'.format)
purchase_df
```

# Gender Demographics

```python
gender_df = purchase_data_df.loc[:,["Gender","SN"]]
gender_df = gender_df.drop_duplicates("SN")

count = gender_df["Gender"].value_counts()
Total_player = len(purchase_data_df["SN"].value_counts())

#Percentage and Count of Male Players
Male = count["Male"]
Male_Percent = ((Male/Total_player)*100)
#Percentage and Count of Female Players
Female = count["Female"]
Female_Percent = ((Female/Total_player)*100)
#Percentage and Count of Other / Non-Disclosed
Other = count["Other / Non-Disclosed"]
Other_Percent = ((Other/Total_player)*100)
#Table Format
gender_df = pd.DataFrame({"": ['Male', 'Female','Other/Non-Disclosed'],
                          "Total Count": [Male, Female, Other],
                          "Percentage of Players": [Male_Percent, Female_Percent, Other_Percent]})
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map('{:,.2f}%'.format)
gender_df = gender_df.set_index('')
gender_df
```

# Purchasing Analysis (Gender)

```python
#The below each broken by gender...Avg Total Purchase per Person
grouped_df = purchase_data_df.groupby(["Gender"])
unique_df = grouped_df.nunique()
unique_count = unique_df["SN"].unique()
#Purchase Count
count = grouped_df["Gender"].value_counts()
#Average Purchase Price
Avg_Purchase_price = grouped_df["Price"].mean()
#Total Purchase Value
Total_Purchase = grouped_df["Price"].sum()
#Average Purchase Total per Person by Gender
Avg_Total_price = Total_Purchase
#Table Format
gender_df = pd.DataFrame({"Gender": ['Female','Male','Other/Non-Disclosed'],
                         "Purchase Count": [count[0],count[1],count[2]],
                         "Average Purchase Price": [Avg_Purchase_price[0], Avg_Purchase_price[1], Avg_Purchase_price[2]],
                         "Total Purchase Value": [Total_Purchase[0], Total_Purchase[1], Total_Purchase[2]],
                         "Avg Total Purchase per Person": [Total_Purchase[0]/unique_count[0], Total_Purchase[1]/unique_count[1], Total_Purchase[2]/unique_count[2]]})
gender_df["Average Purchase Price"] = gender_df["Average Purchase Price"].map("${:,.2f}".format) 
gender_df["Total Purchase Value"] = gender_df["Total Purchase Value"].map("${:,.2f}".format) 
gender_df["Avg Total Purchase per Person"] = gender_df["Avg Total Purchase per Person"].map("${:,.2f}".format) 
gender_df = gender_df.set_index('Gender')
gender_df
```

# Age Demographics

```python
#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_lables = ["<10","10-14", "15-19","20-24","25-29","30-34","35-39","40+"]

Age_df = purchase_data_df.drop_duplicates("SN")
Age_df["Age Group"] = pd.cut(Age_df["Age"], bins, labels=group_lables)
age_group = Age_df.groupby("Age Group")

#Total Count
age_purchase = Age_df["Age Group"].value_counts()
#Percentage of Players
total_player = Age_df["Purchase ID"].count()
Percent_player = (age_purchase/total_player)*100
#Table Format
age_group = pd.DataFrame({"Total Count": age_purchase,
                         "Percentage of Players": Percent_player })
age_group["Percentage of Players"] = age_group["Percentage of Players"].map('{:,.2f}%'.format)
age_group = age_group.reindex(["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])
age_group
```

# Purchasing Analysis (Age)

```python
#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_lables = ["<10","10-14", "15-19","20-24","25-29","30-34","35-39","40+"]

Age1_df = purchase_data_df
Age1_df["Age Group"] = pd.cut(Age1_df["Age"], bins, labels=group_lables)
age1_group = Age1_df.groupby("Age Group")

#Total Count
age1_purchase = Age1_df["Age Group"].value_counts()
#Percentage of Players
avg1_price = age1_group["Price"].mean()
total1_price =age1_group["Price"].sum()
total1_price_person = total1_price / age_purchase

#Table Format
age1_group = pd.DataFrame({"Purchase Count": age1_purchase,
                         "Average Purchase Price": avg1_price,
                         "Total Purchase Value":total1_price,
                         "Avg Total Purchase per Person":total1_price_person})
age1_group["Average Purchase Price"] = age1_group["Average Purchase Price"].map('${:,.2f}'.format)
age1_group["Total Purchase Value"] = age1_group["Total Purchase Value"].map('${:,.2f}'.format)
age1_group["Avg Total Purchase per Person"] = age1_group["Avg Total Purchase per Person"].map('${:,.2f}'.format)
age1_group = age1_group.reindex(["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])
age1_group
```

# Top Spenders

```python
#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

#SN
group_by_sn = purchase_data_df.groupby("SN")
#Purchase Count
age_purchase = group_by_sn["Price"].count()
#Average Purchase Price
avg_price = group_by_sn["Price"].mean()
#Total Purchase Value
total_price = group_by_sn["Price"].sum()
#Table Format
all_spender = pd.DataFrame({"Purchase Count": age_purchase,
                         "Average Purchase Price": avg_price,
                         "Total Purchase Value":total_price})
top_spender = all_spender.sort_values("Total Purchase Value",ascending=False)
top_spender["Average Purchase Price"] = top_spender["Average Purchase Price"].map('${:,.2f}'.format)
top_spender["Total Purchase Value"] = top_spender["Total Purchase Value"].map('${:,.2f}'.format)
top_spender.head()
```

# Most Popular Items

```python
#Identify the 5 most popular items by purchase count, then list (in a table):

#Item ID
group_by_item = purchase_data_df.groupby(["Item ID","Item Name"])
#Item Name
#group_by_name = group_by_item(["Item ID","Item Name"])
#Purchase Count
item_purchase = group_by_item["Price"].count()
#Item Price
item_price = group_by_item["Price"].mean()
#Total Purchase Value
item_purchase_price = group_by_item["Price"].sum()
#Table Format
all_items = pd.DataFrame({"Purchase Count": item_purchase,
                         "Item Price": item_price,
                         "Total Purchase Value":item_purchase_price})
top_items = all_items.sort_values("Purchase Count",ascending=False)
top_items["Item Price"] = all_items["Item Price"].map('${:,.2f}'.format)
top_items["Total Purchase Value"] = all_items["Total Purchase Value"].map('${:,.2f}'.format)
top_items.head()
```

# Most Profitable Items

```python
#Identify the 5 most profitable items by total purchase value, then list (in a table):

#Item ID
group_by_item = purchase_data_df.groupby(["Item ID","Item Name"])
#Purchase Count
item_purchase = group_by_item["Price"].count()
#Item Price
item_price = group_by_item["Price"].mean()
#Total Purchase Value
item_purchase_price = group_by_item["Price"].sum()
#Table Format
all_items = pd.DataFrame({"Purchase Count": item_purchase,
                         "Item Price": item_price,
                         "Total Purchase Value":item_purchase_price})
top_items = all_items.sort_values("Total Purchase Value",ascending=False)
top_items["Item Price"] = all_items["Item Price"].map('${:,.2f}'.format)
top_items["Total Purchase Value"] = all_items["Total Purchase Value"].map('${:,.2f}'.format)
top_items.head()
```
